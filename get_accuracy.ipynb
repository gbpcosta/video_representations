```python
from sklearn.metrics import balanced_accuracy_score
```

```python
all_accs = []
all_accs_adj = []
for file in glob(os.path.join(basedir, '*10*', 'models', 'plt_loss_bkup.h5'), recursive=True):
    print(file)
    for ii, key in enumerate(['crossds_pred_labels_KTH', 'crossds_pred_labels_UCF101', 'crossds_pred_labels_YUP']):
        svm_data = pd.read_hdf(file, key)
        # 'dataset', 'val_labels', 'val_pred'
        
        bacc = []
        bacc_adj = []
        for cl in range(svm_data['val_labels'][ii].shape[1]):
            bal_acc = balanced_accuracy_score(y_true=svm_data['val_labels'][ii][:, cl], y_pred=svm_data['val_pred'][ii][:, cl], adjusted=False)
            bal_acc2 = balanced_accuracy_score(y_true=svm_data['val_labels'][ii][:, cl], y_pred=svm_data['val_pred'][ii][:, cl], adjusted=True)
    
            bacc.append(bal_acc)
            bacc_adj.append(bal_acc2)
            
        all_accs.append({'model_name': file.split('/')[-3], 'val_dataset': svm_data['dataset'], 'bal_acc': bacc, 'bal_acc_adj': bacc_adj})
    
all_accs = pd.DataFrame(all_accs)
all_accs['val_dataset'] = all_accs['val_dataset'].apply(lambda x: x.values[0])
all_accs = all_accs.sort_values(by=['model_name', 'val_dataset'])
all_accs = all_accs.set_index(['model_name', 'val_dataset'])
```
